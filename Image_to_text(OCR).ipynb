```python
# Step 2: Import libraries
import cv2
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

# Step 3: Set up Tesseract path (optional for Colab)
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# Step 4: Function to extract text from an image
def extract_text_from_image(image_path):
    img = cv2.imread(image_path)

    # Optional: Preprocessing
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # OCR using pytesseract
    text = pytesseract.image_to_string(thresh)
    return text

# Step 5: Upload and test
from google.colab import files
uploaded = files.upload()

# Step 6: Extract text from uploaded image
for image_name in uploaded.keys():
    print(f"\n📷 Extracted Text from {image_name}:\n")
    print(extract_text_from_image(image_name))
```
